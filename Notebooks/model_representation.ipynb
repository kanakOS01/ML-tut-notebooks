# Model Representation
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1, 2])
y_train = np.array([300, 500])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
m = x_train.shape[0]
print(f"Number of training examples = {m}")
```

```python
plt.scatter(x_train, y_train, marker='o', c='r')
plt.title('Housing Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price (1000 USD)')
plt.show()
```

## Model Function
<hr>

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, c='r', marker='x', label='Actual Values')

plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (1000 USD)")

plt.legend()
plt.show()
```
